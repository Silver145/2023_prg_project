```python
import turtle
import time
import random

delay = 0.1

# Starting food type
fruit_type = "apple"

# Timers
timer = 150
magic_food_timer = 50

# Food count
food_count = 0

# Score
score = 0
high_score = 0

# Segments
segments = []

# Set up the screen
wn = turtle.Screen()
wn.title("Snake Game")
wn.bgcolor("black")
wn.setup(width=600, height=600)
wn.tracer(0) # Turns off the screen updates

# Snake head
head = turtle.Turtle()
head.speed(0)
head.shape("square") 
head.color("blue")
head.shapesize(stretch_wid=1.1, stretch_len=1.1, outline=1)
head.penup()
head.goto(0,0)
head.direction = "stop"

# Snake apple
apple = turtle.Turtle()
apple.speed(0)
apple.shape("square")
apple.color("red")
apple.penup()
apple.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)

# Snake pear
pear = turtle.Turtle()
pear.speed(0)
pear.shape("square")
pear.color("green")
pear.penup()
pear.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

# Snake banana
banana = turtle.Turtle()
banana.speed(0)
banana.shape("square")
banana.color("yellow")
banana.penup()
banana.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

# Snake watermelon
wm = turtle.Turtle()
wm.speed(0)
wm.shape("square")
wm.color("lime")
wm.penup()
wm.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

# Snake pumpkin
pumpkin = turtle.Turtle()
pumpkin.speed(0)
pumpkin.shape("square")
pumpkin.color("orange")
pumpkin.penup()
pumpkin.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

# Snake pineapple
pa = turtle.Turtle()
pa.speed(0)
pa.shape("square")
pa.color("purple")
pa.penup()
pa.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

# Mmgic food
magic_food = turtle.Turtle()
magic_food.speed(0)
magic_food.shape("square")
magic_food.color("brown")
magic_food.penup()
magic_food.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)

# Text
text = turtle.Turtle()
text.speed(0)
text.shape("square")
text.color("white")
text.penup()
text.hideturtle()
text.goto(0, 260)
text.write("Score: 0  High Score: 0  Food count: 0, Food Timer: 150", align="center", font=("Arial", 14, "normal"))

# Functions
def go_up():
    if head.direction != "down":
        head.direction = "up"

def go_down():
    if head.direction != "up":
        head.direction = "down"

def go_left():
    if head.direction != "right":
        head.direction = "left"

def go_right():
    if head.direction != "left":
        head.direction = "right"

def move():
    if head.direction == "up":
        y = head.ycor()
        head.sety(y + 20)

    if head.direction == "down":
        y = head.ycor()
        head.sety(y - 20)

    if head.direction == "left":
        x = head.xcor()
        head.setx(x - 20)

    if head.direction == "right":
        x = head.xcor()
        head.setx(x + 20)

# Keyboard bindings
wn.listen()
wn.onkeypress(go_up, "w")
wn.onkeypress(go_down, "s")
wn.onkeypress(go_left, "a")
wn.onkeypress(go_right, "d")

# Main game loop
while True:
    wn.update()
    
    # Decrease timer
    timer -= 1
    
    text.clear()
    text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 

    # Check if timer has run out
    if timer == 0:
        score -= 5  # decrease score
        timer = 150  # reset timer
        if score < 0:
            score = 0
            
        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 
    
    # Decrease rotten food timer
    magic_food_timer -= 1
    
    # Check rotten food timer
    if magic_food_timer == 0:
        magic_food.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)
        magic_food_timer = 50
    
    # Change fruit type to pear
    if score >= 50 and fruit_type == "apple":
        pear.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)
        apple.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        fruit_type = "pear"
        
    # Change fruit type to banana
    if score >= 150 and fruit_type == "pear":
        banana.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)
        pear.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        fruit_type = "banana"
        
    # Change fruit type to watermelon
    if score >= 300 and fruit_type == "banana":
        wm.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)
        banana.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        fruit_type = "watermelon"
        
    # Change fruit type to pumpkin
    if score >= 600 and fruit_type == "watermelon":
        pumpkin.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)
        wm.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        fruit_type = "pumpkin"
    
    # Change fruit type to pineapple
    if score >= 1500 and fruit_type == "pumpkin":
        pa.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)
        pumpkin.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        fruit_type = "pineapple"
        
    # Check for a collision with the border
    if head.xcor()>290 or head.xcor()<-290 or head.ycor()>290 or head.ycor()<-290:
        time.sleep(1)
        head.goto(0,0)
        head.direction = "stop"

        # Hide the segments
        for segment in segments:
            segment.goto(1000, 1000)
        
        # Clear the segments list
        segments.clear()

        # Reset the score
        score = 0

        # Reset the delay
        delay = 0.1
        
        # Reset fruit type
        fruit_type = "apple"
        
        # Reset food count
        food_count = 0
        
        # Reset timer
        timer = 150
        
        # Reset rotten food timer
        magic_food_timer = 50
        
        # pineapple go bye bye
        pa.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        
        # pumpkin go bye bye
        pumpkin.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        
        # watermelon go bye bye
        wm.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        
        # Banana go bye bye
        banana.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        
        # Pear go bye bye
        pear.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

        # Apple goes to a new spot
        apple.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)
        
        # Rotten food goes to a new spot
        magic_food.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)
        
        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 

    # Check for a collision with the apple
    if head.distance(apple) < 20:
        # Move the food to a random spot
        x = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        y = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        apple.goto(x,y)

        # Add a segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("blue")
        new_segment.shapesize(stretch_wid=0.7, stretch_len=0.7, outline=1)
        new_segment.penup()
        segments.append(new_segment)

        # Shorten the delay
        delay -= 0.001
        
        # Reset timer
        timer = 150
        
        # Food count increase
        food_count += 1

        # Increase the score
        score += 1

        if score > high_score:
            high_score = score
        
        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 
        
    # Check for a collision with the pear
    if head.distance(pear) < 20:
        # Move the food to a random spot
        x = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        y = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        pear.goto(x,y)

        # Add a segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("blue")
        new_segment.shapesize(stretch_wid=0.7, stretch_len=0.7, outline=1)
        new_segment.penup()
        segments.append(new_segment)

        # Shorten the delay
        delay -= 0.001
        
        # Reset timer
        timer = 150
        
        # Food count increase
        food_count += 1

        # Increase the score
        score += 3

        if score > high_score:
            high_score = score
        
        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 

    # Check for a collision with the banana
    if head.distance(banana) < 20:
        # Move the food to a random spot
        x = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        y = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        banana.goto(x,y)

        # Add a segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("blue")
        new_segment.shapesize(stretch_wid=0.7, stretch_len=0.7, outline=1)
        new_segment.penup()
        segments.append(new_segment)
        
        # Reset timer
        timer = 150
        
        # Food count increase
        food_count += 1

        # Increase the score
        score += 5

        if score > high_score:
            high_score = score
        
        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 
    
    # Check for a collision with the watermelon
    if head.distance(wm) < 20:
        # Move the food to a random spot
        x = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        y = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        wm.goto(x,y)

        # Add a segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("blue")
        new_segment.shapesize(stretch_wid=0.7, stretch_len=0.7, outline=1)
        new_segment.penup()
        segments.append(new_segment)
        
        # Reset timer
        timer = 150
        
        # Food count increase
        food_count += 1

        # Increase the score
        score += 7

        if score > high_score:
            high_score = score
        
        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 
        
    # Check for a collision with the pumpkin
    if head.distance(pumpkin) < 20:
        # Move the food to a random spot
        x = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        y = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        pumpkin.goto(x,y)

        # Add a segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("blue")
        new_segment.shapesize(stretch_wid=0.7, stretch_len=0.7, outline=1)
        new_segment.penup()
        segments.append(new_segment)
        
        # Reset timer
        timer = 150
        
        # Food count increase
        food_count += 1

        # Increase the score
        score += 9

        if score > high_score:
            high_score = score
        
        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 
        
    # Check for a collision with the pineapple
    if head.distance(pa) < 20:
        # Move the food to a random spot
        x = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        y = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        pa.goto(x,y)

        # Add a segment
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("square")
        new_segment.color("blue")
        new_segment.shapesize(stretch_wid=0.7, stretch_len=0.7, outline=1)
        new_segment.penup()
        segments.append(new_segment)
        
        # Reset timer
        timer = 150
        
        # Food count increase
        food_count += 1

        # Increase the score
        score += 11

        if score > high_score:
            high_score = score
        
        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 
    
    # Check for a collision with the magic food
    if head.distance(magic_food) < 20:
        if food_count < 5:
            # Add a segment
            new_segment = turtle.Turtle()
            new_segment.speed(0)
            new_segment.shape("square")
            new_segment.color("blue")
            new_segment.shapesize(stretch_wid=0.7, stretch_len=0.7, outline=1)
            new_segment.penup()
            segments.append(new_segment)
            
            # Decrease score
            score -= 5
            
        elif len(segments) > 0 and food_count >= 10:
            # Remove the last segment
            segments[-1].goto(1000, 1000)
            segments.pop()
            
            # Increase the score
            score += 25
        
        # Move the rotten food to a random spot
        x = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        y = random.randint(-14, 14) * 20  # random multiple of 20 within (-290, 290)
        magic_food.goto(x,y)
        
        # "Reset" timer
        timer = 150
        
        # Reset rotten food timer
        magic_food_timer = 50
        
        # Reset food count
        food_count = 0

        if score > high_score:
            high_score = score
        
        if score < 0:
            score = 0

        text.clear()
        text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 

    # Move the end segments first in reverse order
    for index in range(len(segments)-1, 0, -1):
        x = segments[index-1].xcor()
        y = segments[index-1].ycor()
        segments[index].goto(x, y)


    # Move segment 0 to where the head is
    if len(segments) > 0:
        x = head.xcor()
        y = head.ycor()
        segments[0].goto(x,y)

    move()    

    # Check for head collision with the body segments
    for segment in segments:
        if segment.distance(head) < 20:
            time.sleep(1)
            head.goto(0,0)
            head.direction = "stop"
        
            # Hide the segments
            for segment in segments:
                segment.goto(1000, 1000)

            # Clear the segments list
            segments.clear()

            # Reset the score
            score = 0

            # Reset the delay
            delay = 0.1

            # Reset fruit type
            fruit_type = "apple"

            # Reset food count
            food_count = 0

            # Reset rotten food timer
            magic_food_timer = 50
            
            # Reset timer
            timer = 150
            
            # pineapple go bye bye
            pa.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
        
            # pumpkin go bye bye
            pumpkin.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)
            
            # watermelon go bye bye
            wm.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

            # Banana go bye bye
            banana.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

            # Pear go bye bye
            pear.goto(random.randint(1000, 1000) * 20, random.randint(1000, 1000) * 20)

            # Food goes to a new spot
            apple.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)

            # Rotten food goes to a new spot
            rotten_food.goto(random.randint(-14, 14) * 20, random.randint(-14, 14) * 20)

            text.clear()
            text.write("Score: {}  High Score: {}  Food Count: {} Food Timer: {}".format(score, high_score, food_count, timer), align="center", font=("Arial", 14, "normal")) 

    time.sleep(delay)

wn.mainloop()
```
